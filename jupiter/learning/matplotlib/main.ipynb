```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6, 12, 7, 9, 15])
ypoints = np.array([0, 2, 10, 14, 6, 12])

font_title = {"family":"serif", "color":"blue", "size":20}
font_label = {"family":"arial", "color":"red", "size":15}
font_suptitle = {"family":"arial", "color":"green", "size":45}


plt.subplot(1, 2, 1)
plt.title("Wyk 1", fontdict=font_title, loc="left")
plt.plot(xpoints, ypoints, "D-.m", ms=20, mec="#EEEEEE", mfc="hotpink", linewidth = '10.5')

plt.subplot(1, 2, 2)
plt.plot(xpoints, ypoints, "D", ms=20, mec="#EEEEEE", mfc="hotpink", ls = 'dotted', c="#4CAF50", linewidth = '10.5')
#plt.plot(xpoints)
#plt.plot(ypoints)

plt.title("Tytuł wykresu :)", fontdict=font_title, loc="left")
plt.xlabel("oś x", fontdict=font_label)
plt.ylabel("oś y", fontdict=font_label)
plt.grid(axis="x", color="#C55cC5", linestyle="--", linewidth="0.5")

plt.suptitle("Tytul glowny", fontdict=font_suptitle)

plt.show
```

## Creating Bars
With Pyplot, you can use the bar() function to draw bar graphs:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2,1)
plt.bar(x,y, color="red", width=0.75)
plt.title("Default width=0.8")

plt.subplot(1,2,2)
plt.barh(x,y, height=0.50)
plt.title("Default height=0.8")

plt.show()
```

## Histogram
In Matplotlib, we use the hist() function to create histograms.

```python
x=np.random.normal(170, 10, 250)
plt.hist(x, color="green")

plt.show()
```

## Pie Charts
With Pyplot, you can use the pie() function to draw pie charts.
By default the plotting of the first wedge starts from the x-axis and moves counterclockwise.
The value divided by the sum of all values: x/sum(x)

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["k", "hotpink", "blue", "#4CAF50"]

plt.pie(y, labels=mylabels, startangle=90, explode=myexplode, shadow=True, colors=mycolors)
plt.legend(title="Owoce")

plt.show()
```
